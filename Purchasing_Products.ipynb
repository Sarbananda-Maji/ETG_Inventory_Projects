```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

items = json.loads(r)
```

```python
sales={0:{"Customer Name":"name", "Product ID":"ID", "Name":"Pd name", "Quantity":"Qty", "Amount":"amount"}}
del sales[0]
```

```python
import time

n=int(input("Enter how many Products you want to purchase : "))
bill=0
name=str(input("Enter Your Name : "))
for i in range(n):
    print("\nEnter the",i+1,"Product details")
    id_product = str(input("Enter the Product Id : "))
    quant = int(input("Enter the total quantity of the Product : "))
    print("Product Name : ",items[id_product]["Name"])
    print("Price : ", items[id_product]['Price'])
    if(quant>items[id_product]['Qty']):
        print("Available quantity : ",items[id_product]['Qty'])
        amount = (items[id_product]['Price'] * quant) 
        print("Your Amount : ",amount)
        dis=0
        if(quant>=items[id_product]['Discount for Products']):
            dis = (items[id_product]['Price'] * quant)*(items[id_product]['Discount (%)']/100)
            print("Discount : ", dis)
        
        totalAmount=amount-dis
        print("Amount after discount : ", totalAmount)
        bill+=(totalAmount)
        
        sales[len(sales)+1]={'Customer Name':name, 'Product ID':id_product, 'Product Name':items[id_product]["Name"], 'Quantity':quant, 'Amount':totalAmount, 'Date & Time':time.ctime()}
        items[id_product]['Qty'] = items[id_product]['Qty'] - quant
        
    else:
        amount = (items[id_product]['Price'] * quant)
        print("Your Amount : ",amount)
        dis=0
        if(quant>=items[id_product]['Discount for Products']):
            dis = (items[id_product]['Price'] * quant)*(items[id_product]['Discount (%)']/100)
            print("Discount : ", dis)
        
        totalAmount=amount-dis
        print("Amount after discount : ", totalAmount)
        bill+=(totalAmount)
        
        sales[len(sales)+1]={'Customer Name':name, 'Product ID':id_product, 'Product Name':items[id_product]["Name"], 'Quantity':quant, 'Amount':totalAmount, 'Date & Time':time.ctime()}
        items[id_product]['Qty'] = items[id_product]['Qty'] - quant
    
print(" \nBilling Amount : ",bill)
print(" Thank You for Purchasing Please Visit Again")

```

```python
js = json.dumps(items)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
