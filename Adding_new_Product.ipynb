```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

items = json.loads(r)
```

```python
items
```

```python
n=int(input("Enter how many Products you want to add : "))
for i in range(n):
    print("\nEnter the",i+1,"Products details")
    product_Id = str(input("Enter Product Id : "))
    print("Is this Product already available ")
    k=str(input("Enter 'Y'for yes or 'N'for no : "))
    if(k=="Y"):
        print("You want increase the quantity of the Product")
        print("Product Name : ",items[product_Id]['Name'])
        print("Product quantity : ",items[product_Id]['Qty'])
        q=int(input("Enter how much Quantity you want to increase : "))
        items[product_Id]['Qty']= items[product_Id]['Qty']+q
    else:
        name = str(input("Enter Name of the Product : "))
        content = str(input("Enter the Content of the Product : "))
        price = int(input("Enter the Price of the Product : "))
        qty = int(input("Enter Quantity available : "))
        dis_packets = int(input("Enter the no. of Products for the Discount : "))
        discount = int(input("Enter Discount available for this Product(in %) : "))
        
        items[product_Id] = {'Name': name, 'Content': content, 'Price': price, 'Qty': qty, 'Discount for Products':dis_packets, 'Discount (%)': discount}

print("\nThank You for Adding more Products")


js = json.dumps(items)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
